```python
import numpy as np
np.random.seed(100)

x = np.random.normal(0, 10, 100)
x
```

### **Importing the Library**

```python
import matplotlib.pyplot as plt
```

### **Make Histogram Plot**

```python
plt.hist(x) # Use hist() function to make histogram, x is the input data
plt.show() # Display the plot
```

### **Customizing the bar colors**

The **color** input argument allow us to specify the color of the bar.

```python
plt.hist(x, color='green')
plt.show()
```

```python
plt.hist(x, color='skyblue')
plt.show()
```

### **Bin size**

The **bin** input argument allow us to specify how many bins or bars you see in the plot. The total number of bins (bars) is equal to the number that we assign to the bin input argument. A bin represents the interval (or literally the range of values specified at the bin (bar).

At default we can see a *bin size* of **10**.

```python
plt.hist(x)
plt.show()
```

Now, we will set the *bin size* to **5** and so there will be 5 bars.

```python
plt.hist(x, bins=5)
plt.show()
```

### **Title and axis labels**

```python
fig, ax = plt.subplots()

ax.hist(x)
ax.set_title('Histogram plot') # Title
ax.set_xlabel('X axis label')  # X axis label
ax.set_ylabel('Y axis label')  # Y axis label

plt.show()
```

## **Plot styles**

Let's print out the available styles

```python
print(plt.style.available)
```

Let's use a for loop to iterate through all the possible styles

```python
for i in plt.style.available:
  %matplotlib inline
  print(i) # Print style name
  plt.style.use(i) # Style
  plt.hist(x) # Make Histogram
  plt.show()
```

#### **ggplot2**

```python
%matplotlib inline

plt.style.use('ggplot')
plt.hist(x)
plt.show()
```

## **Relative width of bars**

The relative width or **rwidth** can be a value from 0 to 1.

```python
plt.hist(x, rwidth=0.1)
plt.show()
```

```python
plt.hist(x, rwidth=0.5)
plt.show()
```

```python
plt.hist(x, rwidth=0.9)
plt.show()
```

### **Orientation**

By default, the orientation of bars in the plot is **vertical**. Here, we will try setting the value to **'horizontal'**.

```python
plt.hist(x)
plt.show()
```

```python
plt.hist(x, orientation='horizontal')
plt.show()
```

### **Type of histogram bar**

By default, we have the **bar**, here we will try using the **step** value.

```python
plt.hist(x, histtype='step') # 'bar', 'barstacked', 'step', 'stepfilled'
plt.show()
```

### **Edge color of bar**

The *edge color* of bars can be specified by the **ec** input argument.

```python
plt.hist(x, ec = 'black')
plt.show()
```

#### **Figure size**

```python
plt.figure(figsize=(10,5))
plt.hist(x)
plt.show()
```

```python
#Histogram Plot
import pandas as pd
import matplotlib.pyplot as plt
housing_data = pd.read_csv('housing.csv')

plt.rcParams.update({'font.size': 22})
housing_data.hist(grid=True,xlabelsize=20, ylabelsize=25,figsize=(30,30), bins=40 )
plt.show()
```

```python
#scatter plot
plt.rcParams.update({'font.size': 18})
housing_data.plot(kind='scatter',x='longitude',y='latitude',alpha=0.3,
                 c="median_house_value",cmap=plt.get_cmap('jet'),figsize=(15,10))
plt.plot()
```

```python
plt.rcParams.update({'font.size': 12})
from pandas.plotting import scatter_matrix

attributes = ['median_house_value' , 'median_income' , 'total_rooms' , 'housing_median_age']
scatter_matrix(housing_data[attributes], figsize=(12,8))
```

```python
k = 5 #number of variables for heatmap
cols = housing_data.corr().nlargest(k, 'median_income')['median_income'].index
cm = housing_data[cols].corr()
plt.figure(figsize=(14,8))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```

```python
x_line = housing_data["population"].values.reshape(-1,1)
y_line = housing_data["total_rooms"].values.reshape(-1,1)
z_line = housing_data["total_bedrooms"].values.reshape(-1,1)
fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection="3d")
ax.scatter3D(x_line, y_line, z_line, 'gray')
plt.show()
```
