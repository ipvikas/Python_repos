```python
import socket
import ssl

context = ssl.SSLContext(ssl.PROTOCOL_TLSv1)
context.verify_mode = ssl.CERT_REQUIRED
context.check_hostname = True
context.load_default_certs()

server = "pythonprogramming.net"
port = 443
server_ip = socket.gethostbyname(server)

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s = context.wrap_socket(s, server_hostname=server)

request = "GET / HTTP/1.1\nHost: "+server+"\n\n"

s.connect((server,port))
s.send(request.encode())
result = s.recv(4096)

while (len(result) > 0):
    print(result)
    result = s.recv(4096)
```

```python
#BEST IMP
import re
import urllib.request
#for hyperlinks
print("1. Hyperlinks are:")
fhand = urllib.request.urlopen('https://pythonprogramming.net/')  #https://www.python.org/'
for line in fhand:
    line=line.decode('utf-8')
    res_hyperlink = re.finditer(r'(https?://[a-zA-Z0-9._/]+)',line) #for hyperlinks
    if res_hyperlink:        
        for i in res_hyperlink:
            print(i.group())
            
            
            
#for title
print("\n 2. Title is:")
fhand = urllib.request.urlopen('https://www.python.org/') # each time this line is required
for line in fhand:
    line=line.decode('utf-8')
    res_title = re.finditer(r'<title>.*</title>',line) #for title
    if res_title:
        for i in res_title:
            print(i.group())
            
            
#for script type
print("\n 3. Script Type is:")
fhand = urllib.request.urlopen('https://www.python.org/')
for line in fhand:
    line=line.decode('utf-8')
    res_script = re.finditer(r'<script type="(\S+)">',line) #for script type
    if res_script:
        for i in res_script:
            print(i.group())

```

```python
#EXAMPLE #2 IMP

from bs4 import BeautifulSoup
import requests
import csv

source = requests.get('http://coreyms.com').text

soup = BeautifulSoup(source, 'lxml')

csv_file = open('cms_scrape.csv', 'w')

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline', 'summary', 'video_link'])

for article in soup.find_all('article'):
    headline = article.h2.a.text
    print(headline)

    summary = article.find('div', class_='entry-content').p.text
    print(summary)

    try:
        vid_src = article.find('iframe', class_='youtube-player')['src']

        vid_id = vid_src.split('/')[4]
        vid_id = vid_id.split('?')[0]

        yt_link = f'https://youtube.com/watch?v={vid_id}'
    except Exception as e:
        yt_link = None

    print(yt_link)

    print()

    csv_writer.writerow([headline, summary, yt_link])

csv_file.close()
```
